```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from google.colab import drive
drive.mount('/content/drive')
MODEL_PATH = "/content/drive/MyDrive/poemas/modelo_anafora"

# Cargar modelo y tokenizador
tokenizer = AutoTokenizer.from_pretrained(MODEL_PATH)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_PATH,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
)
if torch.cuda.is_available():
    model.cuda()

# FUNCIÓN DE GENERACIÓN
def generar_anafora(palabra):
    prompt = (
        f'Escribe un poema usando la figura retórica "anáfora" '
        f'y usando la palabra "{palabra}".\nPoema:\n'
    )

    # Tokenizar entrada
    inputs = tokenizer(prompt, return_tensors="pt")
    if torch.cuda.is_available():
        inputs = {k: v.cuda() for k, v in inputs.items()}

    # Generación
    output = model.generate(
        **inputs,
        max_new_tokens=150,
        temperature=0.85,        # creatividad suave
        top_p=0.92,              # controla diversidad
        do_sample=True,
        repetition_penalty=1.15  # evita loops
    )

    # Decodificar y mostrar
    texto = tokenizer.decode(output[0], skip_special_tokens=True)
    print(" Poema generado:\n")
    print(texto)
generar_anafora("Corazon")
```
